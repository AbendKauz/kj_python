```python
import pandas as pd
```

```python
data = pd.read_csv("testest.csv", encoding="utf-8")
data
```

```python
data["level"].unique()
```

```python
data = data.astype({"length":"float", "time":"int", "uid":"object"})
```

```python
data.info()
```

```python
data["area"].unique()
```

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
X
```

```python
area_dum = pd.get_dummies(data["area"])
X = pd.concat([X, area_dum], axis=1)
X.drop(["area"], inplace=True, axis=1)
X
```

```python
level_dum = pd.get_dummies(data["level"])
X = pd.concat([X, level_dum], axis=1)
X.drop(["level"], inplace=True, axis=1)
X
```

```python
ID_dum = pd.get_dummies(data["uid"])
X = pd.concat([X, ID_dum], axis=1)
X.drop(["uid"], inplace=True, axis=1)
X
```

```python
keyword_dum = pd.get_dummies(data["keyword"])
X = pd.concat([X, keyword_dum], axis=1)
X.drop(["keyword"], inplace=True, axis=1)
X
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
X = sc.fit_transform(X)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
for i in range(1, 51):
    model = RandomForestClassifier(max_depth=i, random_state=123)
    model.fit(X[:-1], y[:-1])
    X_test = X[-1:]
    y_test = y[-1:]
    print("depth: {}".format(i))
    print(model.score(X_test, y_test))
    
```

```python
model = RandomForestClassifier(max_depth=30, random_state=123)
```

```python
model.fit(X[:-39], y[:-39])
```

```python
X_test = X[-39:]
X_test
```

```python
y_test = y[-39:]
y_test
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
import joblib
```

```python
joblib.dump(model, 'model_v3.pkl')
```

```python
loaded_model = joblib.load('./model_v3.pkl')
```

```python
loaded_model.score(X_test, y_test)
```
