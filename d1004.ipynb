```python
import pandas as pd
# 딕셔너리를 이용한 Series 데이터 생성
dic = {'name':['kim', 'lee', 'heo'], 'kor':[80, 90, 75], 'eng':[80, 95, 100]}
ser_data = pd.Series(dic)
# Series 데이터 생성 
# 형식 -> pd.Series(data [, index = 인덱스])
ser_data
```

```python
# Numpy 배열을 이용한 Series 데이터 생성
import numpy as np
arr = np.array([80, 90, 75])
ser_data = pd.Series(arr)
ser_data
```

```python
type(ser_data)
```

### DataFrame 생성
- 형식 -> pandas.DataFrame(data [, index = 인덱스, columns = 레이블 목록, dtype = 데이터 타입])
- index = 축(행) 레이블의 리스트, 인덱스가 없으면 0-Based index가 기본 사용됨
- columns : 열의 레이블 리스트를 할당, 인덱스가 없으면 0-Based index가 기본 사용됨
- dtype : column마다 데이터 유형을 명기, 인덱스가 없으면 0-Based index가 기본 사용됨

```python
arr = np.array([80, 90, 75])
df = pd.DataFrame(arr)
df
```

```python
type(df)
```

```python
dic = {'name':['kim', 'lee', 'heo'], 'kor':[80, 90, 75], 'eng':[80, 95, 100]}
df = pd.DataFrame(dic)
df
```

####색인 객체

- RDBMS(관계형 데이터베이스)의 기본키와 유사하게 Series 혹은 DataFrame의 데이터를 식별하는 객체

- 형식 : index 객체명 = pd.Index(index list, dtype = None, name = None)
-- index list : 인덱스 이름 목록
-- dtype : 데이터의 유형, None이면 가장 적합한 데이터 유형을 찾음
-- name : 인덱스 객체의 이름


```python
s_id = pd.Index(['name', 'kor', 'eng'])   # 색인객체 생성
s_id
```

```python
s_data = pd.Series([['kim', 'lee', 'kang'], [80, 90, 75], [100, 80, 70]])
s_data
```

```python
s_data.index = s_id   # Series 객체에 색인객체 할당
s_data
```
