◆ DataFrame에 행과 열 추가하기

1) 행 추가하기

DataFrame객체명.loc[새로운 행이름]=값 혹은 배열

```python
daf=pd.DataFrame({'name':['kim','lee','kang'],'kor':[90,80,70],'eng':[40,50,60]})
daf                         
```

▶ 이름 : song, 국어 : 95, 영어 : 80 인 행 추가

```python
daf.loc[3]=['song', 95, 80]
daf
```

```python
daf.loc[0]=['park', 75, 70]
# 0번째 행에 덮어쓰기 됨
daf
```

2) 열 추가하기

DataFrame객체명[새로운 열이름]=값 혹은 배열

▶ 수학점수 70,85,90을 추가하기

```python
daf['math']=[70, 85, 90, 70]
daf
```

```python
daf.loc[0]=['kim', 90, 40, 70]
daf
```

◆ 행과 열 삭제하기

1) 행 삭제하기

DataFrame객체명.drop(행 인덱스 또는 배열, axis=0, inplace=True 또는 False)

- axis = 0 이면 행, 1이면 열

- inplace = True 이면 원 객체에 적용 / False이면 원 객체를 변경하지 않고 변경된 새로운 DataFrame 생성


```python
daf.drop(0, axis=0, inplace=False) # 원 객체가 변경되지 않음
```

```python
daf
```

```python
daf.drop(0, axis=0, inplace=True) # 원 객체가 변경됨
daf
```

2) 열 삭제하기

DataFrame객체명.drop(열 인덱스 또는 배열, axis=1, inplace=True 또는 False)


```python
daf.drop('math', axis=1, inplace=False)
```

```python
daf
```

◆ 행 선택하기

1) loc 를 이용한 행 선택

- 1개의 행 선택

DataFrame객체명.loc[선택할 행이름]

- 범위지정 선택

DataFrame객체명.loc[시작행이름:끝행이름]
=> 범위의 끝을 포함함

```python
df=pd.DataFrame({'name':['kim','lee','kang'],'kor':[90,80,70],'eng':[40,50,60]})
df
```

ex) 첫번째 행 선택

```python
first=df.loc[0]
first
```

```python
first=df.loc[2]
first
```

```python
df.loc[0:2]
```

```python
two = df.loc[0:1]
two
```

2) iloc를 이용한 행 선택

- 1개의 행 선택

DataFrame객체명.iloc[선택할 인덱스]

- 범위지정 선택

DataFrame객체명.iloc[시작인덱스:끝인덱스:슬라이싱간격]
=> 범위의 끝을 제외함 / 슬라이싱 간격은 생략할 수 있으며, 생략할 경우 1로 간주함
/ 슬라이싱 간격 -1로 지정하면 역순 제시 / 모든 행을 선택할 경우 시작인덱스와 끝인덱스 지정하지 않음

```python
df
```

```python
df.iloc[1]
```

```python
df.loc[0:2]   # loc : 0 ~ 2까지
```

```python
df.iloc[0:2]   # iloc : 0 ~ 2 바로 앞(1)까지
```
