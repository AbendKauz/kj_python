통계함수

1) 평균값 : mean
    
2) 중간값(중앙값, 중위수) : median
    
3) 최빈값 : mode
    
4) 최소값 : min

5) 최대값 : max

```python
import pandas as pd
```

1) 평균값

[형식] DataFrame객체.mean(axis=None, skipna=None)

  - axis에 설정된 행(1) 또는 열(0)의 값들의 평균값을 계산(기본값 0)
    
  - skipna에는 Na/null값을 제외할 것인지 설정(기본값 True(None제외))

▶ 주어진 DataFrame 객체의 열 값들의 평균값 구하기(None 제외)

```python
df=pd.DataFrame({'X':[1,2,None,3],
                'Y':[4,3,8,4]})
```

```python
df
```

```python
df.mean() # axis 기본값 0 = 행별로 계산
```

```python
df.mean(axis=1) # axis 1 = 열별로 계산
```

```python
df.mean(axis=1,skipna=False) # skipna=False : None/null 포함 계산
```

◆ 상관계수

- 두 개의 연속형 변수(열) 간의 연관성에 대한 측도
- 종류 : 피어슨(pearson), 켄달(kendall), 스피어만(spearman)
    
1) pearson 상관계수

- 가장 많이 활용됨

▶ 상관계수 구하기

[형식] DataFrame객체.corr(method='pearson', min_periods=1)

- method에는 상관의 종류 명기

- min_periods는 유효한 결과를 얻기 위해 열 쌍당 필요한 최소 관측치 수

● 국어와 영어 성적 간의 피어슨 상관계수 구하기

```python
df=pd.DataFrame({'name':['kim','lee','kang'], 'kor':[80,90,75], 'eng':[80,95,100]})
df
```

```python
df.corr() # 1에 가까울수록 일치함을 의미
```

```python
df.corr(method='pearson')
```

```python
df.corr(method='spearman')
```

```python
df.corr(method='kendall')
```
