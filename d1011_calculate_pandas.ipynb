 1) Series 간 연산 규칙
    
    (1) Series 간 연산 방법
       
        - 두 Series 객체 간의 연산은 동일한 index끼리 연산 수행
        - Series 객체의 크기 또는 index가 일치하지 않는 경우 연산 불가
        
    (2) 크기가 다른 경우
       
        - 존재하지 않는 index는 NaN 으로 채움
        

    (3) 두 Series 객체의 덧셈 연산 수행하기(동일크기)
       

```python
score1=pd.Series({'kor':80,'eng':90,'math':85})
score2=pd.Series({'eng':70,'kor':75,'math':85})
res=score1+score2
print(res)
```

    (4) 두 Series 객체의 덧셈 연산 수행하기(다른 크기)
    
       - 존재하지 않는 index는 NaN으로 채움

```python
score1=pd.Series({'kor':80,'eng':90,'math':85})
score2=pd.Series({'eng':70,'kor':75})
res=score1+score2
print(res)
```

 2) Series 사칙 연산 함수
    
    (1) 사칙연산 함수
       
        - 덧셈 : add, 뺄셈 : sub, 곱셈 : mul, 나눗셈 : div
                        
    (2) 두 객체 사이에 공통된 인덱스가 없거나 NaN이 포함된 경우, NaN을 반환
       
    (3) NaN이 반환되는 것을 방지 : fill_value 

    (4) 서로 크기가 다른 두 Series 객체의 덧셈 연산 수행하기(add함수 이용)

```python
score1=pd.Series({'kor':80,'eng':90,'math':85})
score2=pd.Series({'eng':70,'kor':75})
res=score1.add(score2, fill_value=0)
print(res)
```

 3) 그룹(분할) 객체 연산하기
    
    (1) 데이터 집계
    
      - 집계 함수를 이용하여 그룹별 통계를 낼 수 있음
            
      - mean, max, min, sum, count, size, var 등
    
      - 그룹화한 열을 제외한 모든 필드에 집계 함수 적용
        
      - 특정 열에 집계 함수 적용도 가능
    
      - [함수형식] DataFrame객체.groupby(그룹화할 열).집계함수()

      - seaborn의 titanic 데이터셋에서 'survived','sex','age','fare'를 추출하여  
      생존여부 내 성별로 나이와 요금의 평균을 구하기

```python
titan=sns.load_dataset('titanic')
df=titan.loc[7:16, ['survived','sex','age','fare']]
group_res=df.groupby(['survived','sex'])
res=group_res.mean()
print(res)
```

      - seaborn의 titanic 데이터셋에서 'survived','sex','age','fare'를 추출하여 생존여부 내 성별로 나이의 평균만 구하기

```python
titan=sns.load_dataset('titanic')
df=titan.loc[7:16, ['survived','sex','age','fare']]
group_res=df.groupby(['survived','sex'])
res=group_res.mean()['age']
print(res)
```

    (2) agg 함수
    
      - 집계 연산을 포함한 사용자정의함수를 그룹 객체에 적용
            
      - 여러 개의 함수를 사용하여 데이터 집계 연산 처리 가능
    
      - 모든 열에 여러 개의 함수를 적용하거나, 열마다 서로 다른 함수를 적용하는 것도 가능
        
      - 집계 연산을 포함한 여러 함수 매핑 : group객체.agg(사용자정의함수)
    
      - 모든 열에 여러 함수 매핑 : group객체.agg([함수1, 함수2,....])
            
      - 열마다 서로 다른 함수 매핑 : group객체.agg({'열명1':함수1, '열명2':함수2,...})

▶ seaborn의 titanic 데이터셋에서 'survived','sex','age','fare'를 추출하여 생존 여부내에서 성별로 평균-최소값을 구하기

 - 이때, 평균에서 최소값을 빼는 사용자정의함수를 이용

```python
import seaborn as sns
def mean_min(x):
    return x.mean()-x.min()
titan=sns.load_dataset('titanic')
df=titan.loc[7:16,['survived','sex','age','fare']]
group_res=df.groupby(['survived','sex'])
mean_min_res=group_res.agg(mean_min)
print(mean_min_res)
```

▶ seaborn의 titanic 데이터셋에서 'survived','sex','age','fare'를 추출하여 생존 여부내에서 성별로 평균-최소값을 구하기

```python
import seaborn as sns
titan=sns.load_dataset('titanic')
df=titan.loc[7:16,['survived','sex','age','fare']]
group_res=df.groupby(['survived','sex'])
mean_min_res=group_res.agg(['mean','min'])
print(mean_min_res)
```

▶ seaborn의 titanic 데이터셋에서 'survived','sex','age','fare'를 추출하여 생존 여부내에서 성별로 나이의 평균, 요금의 평균과 최소값 구하기

```python
import seaborn as sns
titan=sns.load_dataset('titanic')
df=titan.loc[7:16,['survived','sex','age','fare']]
group_res=df.groupby(['survived','sex'])
mean_min_res=group_res.agg({'age':'mean','fare':['mean','min']})
print(mean_min_res)
```
