2. 함수 매핑하기

  1) 함수 매핑이란?
    
    (1) Series 또는 DataFrame의 개별 원소를 특정
     
        - 함수에 일대일로 대응시키는 과정
        
    (2) 사용자가 직접 만든 함수 매핑 가능
     
        - pandas 기본 함수로 처리하기 어려운 복잡한 연산 적용 가능
 

  2) Series 원소에 함수 매핑하기
    
    (1) 개별 원소 함수 매핑이란?
     
        - apply() 함수 이용
        
        - 인자로 전달하는 매핑 함수에 Series의 모든 원소를 하나씩 입력하고 함수의 반환값을 돌려받음
        
        - Series 원소의 개수만큼 반환값을 받아서 같은 그기의 Series 객체로 반환함
          Series객체.apply(매핑 함수)
        
    (2) 개별 원소 함수 매핑 적용
     
        - seaborn 패키지 내 데이터셋 중 'titanic'의 생존 여부(survived),성별(sex), 나이(age), 요금(fare)을 불러와서 사용자 함수를 적용해 봄

  ① 정의된 사용자 함수 매핑하기

```python
# titanic 데이터셋을 읽기 위해서 seaborn을 import함
import seaborn as sns
# 사용자 함수를 만듬
def minus(data1, data2):
    return data1-data2
# titanic 데이터셋을 읽어서 titan 변수에 저장
titan=sns.load_dataset('titanic')
# titan 데이터셋에서 특정 속성만 추출하여 변수에 저장
df=titan.loc[:,['survived','sex','age','fare']]
#사용자 함수 minus를 매핑함
print(df['fare'].apply(minus,data2=10))
```

 ② lambda 함수 매핑하기

```python
# titanic 데이터셋을 읽기 위해서 seaborn을 import함
import seaborn as sns
# 사용자 함수를 만듬
def minus(data1, data2):
    return data1-data2
# titanic 데이터셋을 읽어서 titan 변수에 저장
titan=sns.load_dataset('titanic')
# titan 데이터셋에서 특정 속성만 추출하여 변수에 저장
df=titan.loc[:,['survived','sex','age','fare']]
#lambda 함수를 이용하여 매핑함
print(df['fare'].apply(lambda x:minus(x,10)))
```

        - applymap() 함수 이용
        
        - 인자로 전달하는 매핑 함수에 DataFrame의 모든 원소를 하나씩 입력하고 함수의 반환값을 돌려받음
        
        - 동일한 형태의 DataFrame 객체로 반환함
          DataFrame객체.applymap(매핑 함수)
        
    (2) 개별 원소 함수 매핑 적용
     
        - seaborn 패키지 내 데이터셋 중 'titanic'의 생존 여부(survived),성별(sex), 나이(age), 요금(fare)을 불러와서 사용자 함수를 적용해 봄 

  ① 정의된 사용자 함수 매핑하기

```python
# titanic 데이터셋을 읽기 위해서 seaborn을 import함
import seaborn as sns
# 사용자 함수를 만듬
def minus(data):
    return data-10
# titanic 데이터셋을 읽어서 titan 변수에 저장
titan=sns.load_dataset('titanic')
# 모든 DataFrame 요소에 연산하므로 수치 데이터만 읽어옴
df=titan.loc[:,['age','fare']]
# 사용자 함수  minus를 매핑함
print(df.applymap(minus))
```

 ② lambda 함수 매핑하기

```python
# titanic 데이터셋을 읽기 위해서 seaborn을 import함
import seaborn as sns
# 사용자 함수를 만듬
def minus(data):
    return data-10
# titanic 데이터셋을 읽어서 titan 변수에 저장
titan=sns.load_dataset('titanic')
# 모든 DataFrame 요소에 연산하므로 수치 데이터만 읽어옴
df=titan.loc[:,['age','fare']]
# lambda 함수를 매핑함
print(df.applymap(lambda x:minus(x)))
```

3. DataFrame에 함수 매핑하기

  1) DataFrame의 각 열에 매핑하기
    
    (1) apply(axis=0) 함수 이용
     
    (2) 모든 열을 하나씩 분리하여 매핑 함수의 인자로 각 열(Series)을 전달
    
    (3) Series를 입력받고, Series를 반환하는 함수를 매핑하면 DataFrame을 반환함
    
    (4) 매핑함수에 따라 반환되는 객체의 종류가 다름
     
        - DataFrame객체.apply(매핑함수, axis=0)
        

    (5) Series를 입력받고 Series를 반환하는 형태
 

```python
# titanic 데이터셋을 읽기 위해서 seaborn을 import함
import seaborn as sns
# 각 요소가 null인지를 판단하여 Series 반환
def value_isnull(data):
    return data.isnull()
# titanic 데이터셋을 읽어서 titan 변수에 저장
titan=sns.load_dataset('titanic')
# 모든 DataFrame 요소에 연산하므로 수치 데이터만 읽어옴
df=titan.loc[:,['age','fare']]
# 사용자 함수를 매핑함
print(df.apply(value_isnull,axis=0))
```

    (6) Series를 입력받고 하나의 값(통계값)으로 반환하는 형태

```python
# titanic 데이터셋을 읽기 위해서 seaborn을 import함
import seaborn as sns
# 평균-최소값
def mean_min(data):
    return data.mean()-data.min()
# titanic 데이터셋을 읽어서 titan 변수에 저장
titan=sns.load_dataset('titanic')
# 모든 DataFrame 요소에 연산하므로 수치 데이터만 읽어옴
df=titan.loc[:,['age','fare']]
# 사용자 함수를 매핑함
print(df.apply(mean_min))
```

  2) DataFrame의 각 행에 매핑하기
    
    (1) apply(axis=1) 함수 이용
     
    (2) 모든 행을 하나씩 분리하여 매핑 함수의 인자로 각 행을 전달
    
    (3) DataFrame의 행 인덱스가 매핑 결과로 반환되는 Series의 인덱스가 됨
    
    (4) 매핑함수에 따라 반환되는 객체의 종류가 다름
     
        - DataFrame객체.apply(매핑함수, axis=1)

    (5) 두 객체를 인자로 전달 받아 뺄셈(-) 연산을 수행해 봄

```python
# titanic 데이터셋을 읽기 위해서 seaborn을 import함
import seaborn as sns
# 두 열의 차를 구하는 사용자 정의 함수
def minus(data1, data2):
    return data1-data2
# titanic 데이터셋을 읽어서 titan 변수에 저장
titan=sns.load_dataset('titanic')
# 모든 DataFrame 요소에 연산하므로 수치 데이터만 읽어옴
df=titan.loc[:,['age','fare']]
# 사용자 함수를 lambda 식으로 매핑함
print(df.apply(lambda x:minus(x['age'],x['fare']),axis=1))
```

  3) DataFrame 객체에 매핑하기
    
    (1) pipe() 함수 이용
     
    (2) 매핑 함수의 반환값에 따라 객체의 종류가 다름
    
    (3) DataFrame만 반환, Series만 반환, 개별 값 반환으로 나뉨
    
        - DataFrame객체.pipe(매핑 함수)    


    (4) DataFrame 만 반환하는 형태

```python
# titanic 데이터셋을 읽기 위해서 seaborn을 import함
import seaborn as sns
# DataFrame 을 전달받아 null 여부를 판단한 DataFrame을 반환
def value_isnull(data):
    return data.isnull()
# titanic 데이터셋을 읽어서 titan 변수에 저장
titan=sns.load_dataset('titanic')
# 모든 DataFrame 요소에 연산하므로 수치 데이터만 읽어옴
df=titan.loc[:,['age','fare']]
# 사용자 함수를 매핑함
print(df.pipe(value_isnull))
```

    (5) Series 만 반환하는 형태

```python
# titanic 데이터셋을 읽기 위해서 seaborn을 import함
import seaborn as sns
# DataFrame 의 각 열의 평균
def col_avg(data):
    return data.mean()
# titanic 데이터셋을 읽어서 titan 변수에 저장
titan=sns.load_dataset('titanic')
# 모든 DataFrame 요소에 연산하므로 수치 데이터만 읽어옴
df=titan.loc[:,['age','fare']]
# 사용자 함수를 매핑함
print(df.pipe(col_avg))
```

    (6) 개별 값을 반환하는 형태

```python
# titanic 데이터셋을 읽기 위해서 seaborn을 import함
import seaborn as sns
# DataFrame 의 각 열의 개수
def col_cnt(data):
    return data.count()
# DataFrame 의 모든 데이터의 총개수
def all_cnt(data):
    return col_cnt(data).sum()
# titanic 데이터셋을 읽어서 titan 변수에 저장
titan=sns.load_dataset('titanic')
# 모든 DataFrame 요소에 연산하므로 수치 데이터만 읽어옴
df=titan.loc[:,['age','fare']]
# 사용자 함수를 매핑함
print(df.pipe(all_cnt))
```
