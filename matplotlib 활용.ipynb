    1) Matplotlib 활용
        (1) 직선 또는 곡선으로 연결하여 데이터 값 사이의 관계를 표시
        (2) 연속적인 값의 변화와 패턴을 파악하는 데 적합
        (3) pyplot 서브패키지를 이용하여 그래프를 만들고 변화를 줌
        (4) subplot 함수를 이용하면, Multi Graph를 그릴 수 있음
        (5) show 함수를 이용하여 그래프를 화면에 나타냄
        (6) savefig 함수를 이용하여 파일로 저장
        (7) 선 그래프 그리기
            - pyplot의 plot 함수 이용
            - [형식] pyplot.plot(x축 값, y축 값, color, linestyle, marker, markerfacecolor, markersize, linewidth)
                - color : 선의 색(HTML color name 또는 Hex Code)
                - linestyle : 선 종류로 사용할 문자
                - marker : 표식으로 사용할 문자
                - markerfacecolor : 표식의 색 지정
                - markersize : 표식의 크기
                - linewidth : 선 굵기
        (8) HTML color name
            - https://www.w3schools.com/colors/colors_names.asp
        (9) Hex Code
            - https://www.w3schools.com/colors/colors_picker.asp
        (10) 차트의 제목 표시 => title(제목)
        (11) 축 이름 표시 => xlabel(x축 이름), ylabel(y축 이름)
        (12) 범례 표시 => legend(labels, loc, fontsize)
            - labels : 항목 레이블 목록
            - loc : 항목 표시 위치 설정
                - 0 : best , 1 : upper right , 2 : upper left , 3 : lower left , 4 : lower right , 5 : right , 6 : center left , 
                7 : center right , 8 : lower center , 9 : lower center , 10 : center
            - fontsize : 레이블 글자 크기
        (13) 차트 크기 : figure(figsize(가로, 세로))
        (14) 축 눈금 라벨 회전 : yticks, xticks
            - rotation : 눈금 레벨 회전(vertical)
            - size : 글자 크기
        (15) y축 숫자 값 범위 지정 : ylim(최소값, 최대값)

```python
from matplotlib import pyplot as plt

from matplotlib import font_manager, rc
font_path='D2Coding-Ver1.3.2-20180524-all.ttc'  # 글꼴 파일 설정
font_name=font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

plt.figure(figsize=(10,5))
#plt.style.use('ggplot')
#plt.style.use('default')
plt.plot(['hong','lee','cho','jeong','kang'],[80,90,100,65,75], color='red', linestyle='--', linewidth=3, marker='o', 
            markerfacecolor='blue')
plt.title('중간고사 성적')
plt.xlabel('이름')
plt.ylabel('점수')
plt.xticks(size=20)
plt.legend(labels=['중간고사 점수'], loc='best')

plt.show()
```

 - 히스토그램 그래프 그리기
    
    - pyplot의 hist 함수 이용
    -[형식] pyplot.hist(x, bins, histtype, facecolor, alpha)
        - x : 그래프 그릴 데이터
        - bins : 히스토그램의 가로축 구간의 개수
        - histtype : 히스토그램 종류(bar, barstacked, step, stepfilled)
        - facecolor : 히스토그램의 색
        - alpha : 투명도

예제)

    from matplotlib import pyplot as plt
    plt.hist([80,90,100,65,75], bins=8, color='red', histtype='bar', facecolor='blue')
    plt.show()

```python
from matplotlib import pyplot as plt
plt.hist([80,90,100,65,75], bins=8, color='red', histtype='bar', facecolor='blue')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
df=pd.DataFrame({
                    'name':{0:'Tom',1:'Laura',2:'Kevin',3:'Julia',4:'Jhon'},
                    'score':{0:80,1:90,2:100,3:65,4:75}                 
                })
sns.lineplot(x='name', y='score', data=df)
```
